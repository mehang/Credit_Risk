```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
C=[0.1,0.01, 0.001, 0.0001]
rocs = [0.681,0.68,0.677, 0.654]
plt.plot(C, rocs, color='green', label='Max iteration = 100')
plt.xlabel('Inverse of regularization(C)')
plt.ylabel('ROC AUC')
plt.legend()
plt.title('Effect of regularization')
plt.show()
```

```python
nodes_150 = [50,400,700,1200]
nodes_380 = [50,400,700,1200]
nodes_672 = [50,400,700,1200]

roc_150 = [0.7859, 0.8185, 0.8164, 0.8213]
roc_380 = [0.8693, 0.9071, 0.9180, 0.9017]
roc_672 = [0.9023, 0.9028,  0.8894, 0.8904]

plt.plot( nodes_150, roc_150, color = 'red', label = '150 nodes, C=0.0001')
plt.plot( nodes_380, roc_380, color = 'green', label = '380 nodes, C=0.0001')
plt.plot( nodes_672, roc_672, color = 'blue', label = '672 nodes, C=0.0001')

plt.xlabel('Max Iteration')
plt.ylabel('ROC AUC')
plt.legend()
plt.title('Effect of max iteration')
plt.show()
```
